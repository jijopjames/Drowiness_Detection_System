In this notebook all our pickles are combined into a single pickle for same group. That is `annotations.pkl` ,`video_info_df.pkl` and `frame_info_df.pkl` is merged into a single pickle file called `merged_df.pkl` for the same group of pickle files. 

```python
import os
import pandas as pd
import pickle
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
annots=[]
frame_infos=[]
video_infos=[]

directory = "/content/gdrive/MyDrive/Project/Data/train/"
new_directory = "/content/gdrive/MyDrive/Project/Data/train2/"

files = os.listdir(directory)
for file in files:
    clean_name = os.path.splitext(file)[0]
    if clean_name.endswith('annotations'):
        annots.append(file)
    if clean_name.endswith('video_info_df'):
        video_infos.append(file)
    if clean_name.endswith('frame_info_df'):
        frame_infos.append(file)

for file in annots:
    clean_name = os.path.splitext(file)[0]
    first_part = clean_name[:-12]
    
    for file2 in frame_infos:
        clean_name2 = os.path.splitext(file2)[0]
        first_part2 = clean_name2[:-14]
        if first_part == first_part2:
            frame_info_df = pd.read_pickle(directory+file2)
            annotation = pd.read_pickle(directory+file)
            if len(frame_info_df) !=len(annotation):
                print(file2, len(frame_info_df))
                print(file, len(annotation))
            else: 
                os.rename(directory+file, new_directory+file)
                os.rename(directory+file2, new_directory+file2)
                drowsiness_series=annotation.loc[:,'drowsiness']
                frame_info_df.loc[:,'drowsiness']=drowsiness_series
                frame_info_df.to_pickle(directory+first_part+'_merged_df.pkl')
```
