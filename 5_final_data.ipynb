In this notebook we combine all merged pickle file in single large dataset we will be used for the processing. 

In this process we generate 4 main pickle files
##    
- `test_set.pkl` : which contain all test set data.
- `training_set.pkl`: which contain all train set data.
- `whole_set.pkl`: which is a combination of train and test set data.
- `whole_set_selected.pkl` : its is same as `whole_set.pkl` without the coloums `blink_no`, `blink_start_frame`, `blink_end_frame`.

```python
import os
import pandas as pd
import numpy as np
import pickle
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
directory = "/content/gdrive/MyDrive/Project/Data/test"
pickles = os.listdir(directory)
pickle_list=[]

for pickle_file in pickles:
    clean_name = os.path.splitext(pickle_file)[0]
    if clean_name.endswith("merged_df"):
        pickle = pd.read_pickle("/content/gdrive/MyDrive/Project/Data/test/"+pickle_file)
        pickle_list.append(pickle)

result = pd.concat(pickle_list)
result.reset_index(inplace=True, drop=True)
result.to_pickle("/content/gdrive/MyDrive/Project/Data/test_set.pkl")
print(result.iloc[0])
```

```python
directory = "/content/gdrive/MyDrive/Project/Data/train"
pickles = os.listdir(directory)
pickle_list2=[]

for pickle_file in pickles:
    clean_name = os.path.splitext(pickle_file)[0]
    if clean_name.endswith("merged_df"):
        pickle = pd.read_pickle("/content/gdrive/MyDrive/Project/Data/train/"+pickle_file)
        pickle_list2.append(pickle)

result2 = pd.concat(pickle_list2)
result2.reset_index(inplace=True, drop=True)
result2.to_pickle("/content/gdrive/MyDrive/Project/Data/training_set.pkl")

result3 = pd.concat([result, result2])
result3.reset_index(inplace=True, drop=True)
result3.to_pickle("/content/gdrive/MyDrive/Project/Data/whole_set.pkl")

result4 = result3.drop(columns=['blink_no', 'blink_start_frame', 'blink_end_frame'])
result4.to_pickle("/content/gdrive/MyDrive/Project/Data/whole_set_selected.pkl")
```
